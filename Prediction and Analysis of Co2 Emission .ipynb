```python
from scipy.stats import skew
# Define a threshold for skewness to decide filling strategy for numerical columns
skewness_threshold = 1  # You can adjust this threshold as needed
# Iterate through each column
for column in df.columns:
  if df[column].dtype in [np.int64, np.float64]:
    col_skewness = skew(df[column].dropna())
    if col_skewness < -skewness_threshold:

      mode = df[column].mode().iloc[0]
      df[column].fillna(mode, inplace=True)
    elif -skewness_threshold <= col_skewness <= skewness_threshold:
      median = df[column].median()
      df[column].fillna(median, inplace=True)
    else:
      mean = df[column].mean()
      df[column].fillna(mean, inplace=True)
  elif df[column].dtype == 'object':
    mode = df[column].mode().iloc[0]
    df[column].fillna(mode, inplace=True)


```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
mse=[]
rmse=[]
mae=[]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Evaluate the model
mselr = mean_squared_error(y_test, y_pred)
r2lr = r2_score(y_test, y_pred)
maelr= mean_absolute_error(y_test, y_pred)


print(f"Mean Squared Error: {mselr}")
print(f"R-squared: {r2lr}")
print(f"mean_absolute_error: {maelr}")
mse.append(mselr)
rmse.append(r2lr)
mae.append(maelr)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# Train a Support Vector Regression model
model = SVR(kernel='linear')  # You can try different kernels like 'linear', 'poly', or 'rbf'
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the SVR model
msesvr = mean_squared_error(y_test, y_pred)
r2svr  = r2_score(y_test, y_pred)
maevr= mean_absolute_error(y_test, y_pred)

print(f"mean_absolute_error: {maevr}")
print(f"Mean Squared Error (MSE): {msesvr}")
print(f"R-squared (R2): {r2svr}")
mse.append(msesvr)
rmse.append(r2svr)
mae.append(maevr)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=42)  # You can adjust hyperparameters as needed
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the Decision Tree Regression model
msedtr = mean_squared_error(y_test, y_pred)
r2dtr = r2_score(y_test, y_pred)
maedtr = mean_absolute_error(y_test, y_pred)

print(f"mean_absolute_error: {maedtr}")
print(f"Mean Squared Error (MSE): {msedtr}")
print(f"R-squared (R2): {r2dtr}")

mse.append(msedtr)
rmse.append(r2dtr)
mae.append(maetr)



```

```python
from sklearn.linear_model import Ridge
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# Train a Ridge Regression model
alpha = 1.0
model = Ridge(alpha=alpha)
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the Ridge Regression model
mser = mean_squared_error(y_test, y_pred)
r2r = r2_score(y_test, y_pred)
maer = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mser}")
print(f"R-squared (R2): {r2r}")
print(f"mean_absolute_error: {maer}")
mse.append(mser)
rmse.append(r2r)
mae.append(maer)
```

```python
from sklearn.linear_model import Lasso
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# Train a Lasso Regression model
alpha = 1.0  # You can adjust the regularization strength (alpha)
model = Lasso(alpha=alpha)
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the Lasso Regression model
msel = mean_squared_error(y_test, y_pred)
r2l = r2_score(y_test, y_pred)
mael = mean_absolute_error(y_test, y_pred)

print(f"mean_absolute_error: {mael}")
print(f"Mean Squared Error (MSE): {msel}")
print(f"R-squared (R2): {r2l}")
mse.append(msel)
rmse.append(r2l)
mae.append(mael)
```

```python
model_names=[LinearRegression, SVR, DecisionTreeRegressor, Ridge, Lasso]

plt.figure(figsize=(10, 6))

bar_width = 0.25
index = np.arange(len(model_names))

plt.bar(index, mse, bar_width, label='MSE', alpha=0.7)
plt.bar(index + bar_width, rmse, bar_width, label='RMSE', alpha=0.7)
plt.bar(index + 2 * bar_width, mae, bar_width, label='MAE', alpha=0.7)
plt.xlabel('Regression Models')
plt.ylabel('Error Value')
plt.title('MSE, MAE and RMSE for Different Regression Models')
plt.xticks(index + bar_width / 2, model_names, rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
